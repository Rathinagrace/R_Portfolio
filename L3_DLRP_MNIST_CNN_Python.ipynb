```python
# =============================================================
# CNN DEMONSTRATION: MNIST DIGIT CLASSIFICATION
# Goal: Achieve 99%+ Accuracy using Convolutional Layers
# Platform: Google Colab (Python)
# =============================================================

import os
os.environ["KERAS_BACKEND"] = "tensorflow"
import keras
from keras import layers
import numpy as np

# 1. Load and Prepare Data (Spatial 28x28 format)
(x_train, y_train), (x_test, y_test) = keras.datasets.mnist.load_data()

# Reshape to (Batch_Size, Height, Width, Channels)
x_train = x_train.astype("float32") / 255
x_test = x_test.astype("float32") / 255
x_train = np.expand_dims(x_train, -1)
x_test = np.expand_dims(x_test, -1)

# One-hot encode labels
y_train = keras.utils.to_categorical(y_train, 10)
y_test = keras.utils.to_categorical(y_test, 10)

# 2. Build the CNN Model (Slide Architecture)
model = keras.Sequential([
    keras.Input(shape=(28, 28, 1)),
    layers.Conv2D(32, kernel_size=(3, 3), activation="relu"),
    layers.MaxPooling2D(pool_size=(2, 2)),
    layers.Conv2D(64, kernel_size=(3, 3), activation="relu"),
    layers.MaxPooling2D(pool_size=(2, 2)),
    layers.Flatten(),
    layers.Dropout(0.5),
    layers.Dense(10, activation="softmax"),
])

# 3. Compile
model.compile(loss="categorical_crossentropy", optimizer="adam", metrics=["accuracy"])

# 4. Train
print("\n--- Training CNN (Expect 99% Accuracy) ---")
model.fit(x_train, y_train, batch_size=128, epochs=10, validation_split=0.1)

# 5. Evaluate
score = model.evaluate(x_test, y_test, verbose=0)
print(f"\nTest Loss: {score[0]:.4f}")
print(f"Test Accuracy: {score[1]*100:.2f}%")
```
